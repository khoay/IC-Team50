```python
import pandas as pd


df = pd.read_csv('FF_SR_ data.csv')
#all of these rows that have an empty value for this it drops
df_cleaned = df.dropna(subset=["FF Min", "FF Max", "SR Mean per 100g", "SR Num_Data_pts"])
#replaces any remaining empty values with 0
df_cleaned.fillna(0, inplace=True)
#drops redundant information columns
df_cleaned = df_cleaned.drop(["SR Food description", "SR_Component", "SR_NDB", "Nutrient_id"], axis=1)

#mapping food_category_id 
mapping_df = pd.read_csv('food_category_id table.csv')

mapping_dict = mapping_df.set_index('food_category_id')['description'].to_dict()

# Replace values in the 'food_category_id' column using the map function
df_cleaned['food_category_id'] = df_cleaned['food_category_id'].map(mapping_dict)

column_order = ['FF_NDB', 'rank', 'FF_Component', 'food_category_id', 'FF Food description', 'unit_name', 'SR Mean per 100g',
                 'SR Min', 'SR Max', ' Std_Error', 'SR Num_Data_pts', 'FF Mean per 100g', 'FF Min', 'FF Max', 'FF Median',
                 'FF data_points', 'FF Publication Date']

df_cleaned = df_cleaned[column_order]

df_cleaned = df_cleaned.rename(columns={'  Std_Error': 'Std_Error'})


df_sorted = df_cleaned.sort_values(by='rank')




df_cleaned.to_csv('CLEANED_by_Food.csv', index=False)
df_sorted.to_csv('CLEANED_by_Nutrient.csv', index=False)

df_cleaned.head(50)
```
